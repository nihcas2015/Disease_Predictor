```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#dataset 1
df= pd.read_csv(r'C:\Users\nihca\OneDrive\Documents\vscode\edaproject\dataset.csv')
```

```python
df
```

```python
# Listing all the diseases in the dataset

fealog=[]
for i in df.columns:
    if i !="Disease":
        for j in df[i]:
            if j not in fealog:
                fealog.append(j)
fealog
```

```python
#converting symptom columns to allot each disease a column
dftemp=pd.DataFrame({"Disease":df["Disease"]})
for i in fealog:
    dftemp[i]=0
for i in range(len(df)):
    row = df.iloc[i]
    
    for col in df.columns:
        if col != "Disease":
            value = row[col]
            
            if value in dftemp.columns:
                dftemp.iloc[i, dftemp.columns.get_loc(value)] = 1
df=dftemp
df
```

```python
#dropping disease with no symptoms

symptom_columns = df.columns.drop('Disease')
zero_symptoms = (df[symptom_columns].sum(axis=1) == 0)
df = df[~zero_symptoms]
```

```python
# Dropping diseases with less than 2 symptoms
diseases_to_drop = []

for disease in df['Disease'].unique():
    disease_rows = df[df['Disease'] == disease]
    feature_count = (disease_rows.iloc[:, 1:] == 1).sum(axis=1).mean()
    if feature_count < 2:
        diseases_to_drop.append(disease)

if diseases_to_drop:
    df = df[~df['Disease'].isin(diseases_to_drop)]
df
```

```python
#listing top symptoms
symptom_counts = df.iloc[:, 1:].sum().sort_values(ascending=False)


plt.figure(figsize=(12, 8))
top_symptoms = symptom_counts.head(20)
sns.barplot(x=top_symptoms.values, y=top_symptoms.index)
plt.title('Top 20 Most Common Symptoms in Dataset')
plt.xlabel('Count')
plt.ylabel('Symptom')
plt.tight_layout()
plt.show()
```

```python
#plotting heatmap of correlation between symptoms for top 10
correlation_matrix = df.drop('Disease', axis=1).corr()


plt.figure(figsize=(16, 14))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=False, linewidths=0.5)
plt.title('Correlation Heatmap of Symptoms', fontsize=16)
plt.tight_layout()
plt.show()


mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
tri_df = correlation_matrix.mask(mask)


corr_pairs = tri_df.stack().reset_index()
corr_pairs.columns = ['Symptom 1', 'Symptom 2', 'Correlation']


corr_pairs['Abs_Correlation'] = corr_pairs['Correlation'].abs()
top_corr = corr_pairs.sort_values('Abs_Correlation', ascending=False).head(10)

print("Top 20 Most Correlated Symptom Pairs:")
top_corr[['Symptom 1', 'Symptom 2', 'Correlation']]
```
