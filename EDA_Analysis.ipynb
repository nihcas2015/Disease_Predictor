```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#dataset 1
df= pd.read_csv(r'C:\Users\nihca\OneDrive\Documents\vscode\edaproject\dataset.csv')
df
```

```python
# Listing all the diseases in the dataset

fealog=[]
for i in df.columns:
    if i !="Disease":
        for j in df[i]:
            if j not in fealog:
                fealog.append(j)
fealog
```

```python
#converting symptom columns to allot each disease a column
dftemp=pd.DataFrame({"Disease":df["Disease"]})
for i in fealog:
    dftemp[i]=0
for i in range(len(df)):
    row = df.iloc[i]
    
    for col in df.columns:
        if col != "Disease":
            value = row[col]
            
            if value in dftemp.columns:
                dftemp.iloc[i, dftemp.columns.get_loc(value)] = 1
df=dftemp
df
```

```python
#dropping disease with no symptoms

symptom_columns = df.columns.drop('Disease')
zero_symptoms = (df[symptom_columns].sum(axis=1) == 0)
df = df[~zero_symptoms]
```

```python
# Dropping diseases with less than 2 symptoms
diseases_to_drop = []

for disease in df['Disease'].unique():
    disease_rows = df[df['Disease'] == disease]
    feature_count = (disease_rows.iloc[:, 1:] == 1).sum(axis=1).mean()
    if feature_count < 2:
        diseases_to_drop.append(disease)

if diseases_to_drop:
    df = df[~df['Disease'].isin(diseases_to_drop)]
df
```

```python
#listing top symptoms
symptom_counts = df.iloc[:, 1:].sum().sort_values(ascending=False)


plt.figure(figsize=(12, 8))
top_symptoms = symptom_counts.head(20)
sns.barplot(x=top_symptoms.values, y=top_symptoms.index)
plt.title('Top 20 Most Common Symptoms in Dataset')
plt.xlabel('Count')
plt.ylabel('Symptom')
plt.tight_layout()
plt.show()
```

```python
#listing top 10 diseases
disease_counts = df['Disease'].value_counts().nlargest(10)
        

plt.figure(figsize=(12, 6))
sns.barplot(x=disease_counts.index, y=disease_counts.values, palette='viridis')
plt.title('Top 10 Diseases by Count', fontsize=16)
plt.xlabel('Disease', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

        
plt.show()
        
```

```python
#listing low frequency diseases
disease_counts = df['Disease'].value_counts().nsmallest(10)
        
plt.figure(figsize=(6, 6))
sns.barplot(x=disease_counts.index, y=disease_counts.values, palette='viridis')
plt.title('Last 10 Diseases by Count', fontsize=16)
plt.xlabel('Disease', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Using pivot table to create a disease-symptom relationship matrix
top_n_symptoms = 15
disease_counts = df['Disease'].value_counts()

top_symptoms = symptom_counts.head(top_n_symptoms).index
top_symptoms = top_symptoms[1:]

disease_symptom_matrix = df.groupby('Disease')[top_symptoms].mean()

top_diseases = disease_counts.head(15).index
disease_symptom_heatmap = disease_symptom_matrix.loc[top_diseases]

plt.figure(figsize=(14, 10))
sns.heatmap(disease_symptom_heatmap, cmap='YlOrRd', annot=True, fmt='.2f', linewidths=0.5)
plt.title('Disease-Symptom Relationship Matrix', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Finding top 10 most correlated symptoms using heatmap
correlation_matrix = df.drop('Disease', axis=1).corr()


mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
tri_df = correlation_matrix.mask(mask)

corr_pairs = tri_df.stack().reset_index()
corr_pairs.columns = ['Symptom 1', 'Symptom 2', 'Correlation']

corr_pairs['Abs_Correlation'] = corr_pairs['Correlation'].abs()
top_corr = corr_pairs.sort_values('Abs_Correlation', ascending=False).head(10)

top_features = list(set(top_corr['Symptom 1'].tolist() + top_corr['Symptom 2'].tolist()))

top_corr_matrix = correlation_matrix.loc[top_features, top_features]

plt.figure(figsize=(12, 10))
sns.heatmap(top_corr_matrix, cmap='coolwarm', annot=True, fmt='.2f', linewidths=0.5)
plt.title('Top 10 Most Correlated Symptoms', fontsize=16)
plt.tight_layout()
plt.show()
```
